```python
import re


def extract_retweet_info(message):
    retweet_pattern = re.compile(r"^RT @(\w+) (.+)$")
    match = retweet_pattern.match(message)

    if match:
        original_author = match.group(1)
        original_message = match.group(2)
        return original_author, original_message

    return None, None


# Example usage:
tweet = "RT @original_author This is the original message"
r = extract_retweet_info(tweet)

print("R: ", r)
```

```python
message_rr = df_all["message"].apply(extract_retweet_info)

result_df = pd.DataFrame(message_rr.tolist(), columns=["og_author", "og_message"])
```
