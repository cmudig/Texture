```python
import pandas as pd
```

```python
# NOTE: this file is like 17gb so dont load in memory all at once
path = "../datasets/local/wmt18-data/data"
```

```python
max_lines = 500000
i = 0

english_lines = []
german_lines = []
scores = []

with open(path, mode="r") as f:
    for line in f:
        if i > max_lines:
            break
        i += 1

        try:
            items = line.split("\t")
            en = items[0]
            de = items[1]
            score = items[2][:-1]  # remove newline

            english_lines.append(en)
            german_lines.append(de)
            scores.append(score)
        except Exception as e:
            print("Failed on: ", line)

df = pd.DataFrame({"en": english_lines, "de": german_lines, "score": scores})
```
