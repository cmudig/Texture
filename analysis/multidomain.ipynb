```python
import pandas as pd
```

```python
def read_file(path, colname="text"):
    with open(path, "r") as f:
        sentences = f.readlines()
    sentences = [s.strip() for s in sentences]
    df = pd.DataFrame(sentences, columns=[colname])
    return df


def process_raw():
    medical_de = read_file("../datasets/local/multidomain/medical/dev.de", "de")
    medical_en = read_file("../datasets/local/multidomain/medical/dev.en", "en")

    law_de = read_file("../datasets/local/multidomain/law/dev.de", "de")
    law_en = read_file("../datasets/local/multidomain/law/dev.en", "en")

    medical_all = medical_en.join(medical_de)
    medical_all["domain"] = "medical"

    law_all = law_en.join(law_de)
    law_all["domain"] = "law"

    all_data = pd.concat([medical_all, law_all])

    all_data.to_csv("../datasets/raw/multidomain.csv", index=False)
```
